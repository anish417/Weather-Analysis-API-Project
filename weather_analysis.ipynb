```python
import requests

api_key = "143b3f0d76d9461d8c9144745262406"
city = input("Enter city name: ")

url = f"https://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=yes"

response = requests.get(url)
data = response.json()

print("City:", data["location"]["name"])
print("Temperature:", data["current"]["temp_c"], "°C")
print("Humidity:", data["current"]["humidity"], "%")
print("Wind Speed:", data["current"]["wind_kph"], "kph")
print("Condition:", data["current"]["condition"]["text"])
```

```python
import requests

api_key = "143b3f0d76d9461d8c9144745262406"
city = input("Enter city name: ")

url = f"https://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=yes"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()

    print("City:", data["location"]["name"])
    print("Country:", data["location"]["country"])
    print("Temperature:", data["current"]["temp_c"], "°C")
    print("Condition:", data["current"]["condition"]["text"])
    print("Humidity:", data["current"]["humidity"], "%")
    print("Wind Speed:", data["current"]["wind_kph"], "km/h")
else:
    print("Error:", response.status_code)
```

```python
import requests
import pandas as pd

api_key = "143b3f0d76d9461d8c9144745262406"

cities = ["Hyderabad", "Delhi", "Mumbai", "Chennai", "Bangalore"]

data = []

for city in cities:
    url = f"https://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=yes"

    response = requests.get(url)

    print(f"{city}: {response.status_code}")

    if response.status_code == 200:
        weather = response.json()

        data.append({
            "City": weather["location"]["name"],
            "Temperature": weather["current"]["temp_c"],
            "Humidity": weather["current"]["humidity"],
            "Wind Speed": weather["current"]["wind_kph"]
        })

weather_df = pd.DataFrame(data)

weather_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(weather_df["City"], weather_df["Temperature"])
plt.title("Temperature Comparison Across Cities")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.show()
```
